RXRX1 Results

```python
for i in range(5,105,5):
    for j in range(5,105,5):
        df = pd.read_csv('/home/ubuntu/image-drift-monitoring/data/rxrx1_v1.0/drifted_data/incremental_drift/whylogs_output/profile_compare/rxrx1_{}_v_{}.csv'.format(i,j),delimiter=',',index_col=[0])
        res = []
        for n in range(0,8,1):
            if df.loc[n]['metric'] == 'image.Brightness.mean':
                if df.loc[n]['rating'] == 'Possible Drift (0.05 - 0.15)': res.append(0.15)
                if df.loc[n]['rating'] == 'Detected Drift (0.00 - 0.05)': res.append(1)
                if df.loc[n]['rating'] == 'No Drift (0.15 - 1.00)': res.append(0)
        for n in range(0,8,1):
            if df.loc[n]['metric'] == 'image.Brightness.stddev':                 
                if df.loc[n]['rating'] == 'Possible Drift (0.05 - 0.15)': res.append(0.15)
                if df.loc[n]['rating'] == 'Detected Drift (0.00 - 0.05)': res.append(1)
                if df.loc[n]['rating'] == 'No Drift (0.15 - 1.00)': res.append(0)
        for n in range(0,8,1):
            if df.loc[n]['metric'] == 'image.Hue.mean':                 
                if df.loc[n]['rating'] == 'Possible Drift (0.05 - 0.15)': res.append(0.15)
                if df.loc[n]['rating'] == 'Detected Drift (0.00 - 0.05)': res.append(1)
                if df.loc[n]['rating'] == 'No Drift (0.15 - 1.00)': res.append(0)
        for n in range(0,8,1):
            if df.loc[n]['metric'] == 'image.Hue.stddev':                 
                if df.loc[n]['rating'] == 'Possible Drift (0.05 - 0.15)': res.append(0.15)
                if df.loc[n]['rating'] == 'Detected Drift (0.00 - 0.05)': res.append(1)
                if df.loc[n]['rating'] == 'No Drift (0.15 - 1.00)': res.append(0)
        for n in range(0,8,1):
            if df.loc[n]['metric'] == 'image.Saturation.mean':                 
                if df.loc[n]['rating'] == 'Possible Drift (0.05 - 0.15)': res.append(0.15)
                if df.loc[n]['rating'] == 'Detected Drift (0.00 - 0.05)': res.append(1)
                if df.loc[n]['rating'] == 'No Drift (0.15 - 1.00)': res.append(0)
        for n in range(0,8,1):
            if df.loc[n]['metric'] == 'image.Saturation.stddev':                 
                if df.loc[n]['rating'] == 'Possible Drift (0.05 - 0.15)': res.append(0.15)
                if df.loc[n]['rating'] == 'Detected Drift (0.00 - 0.05)': res.append(1)
                if df.loc[n]['rating'] == 'No Drift (0.15 - 1.00)': res.append(0)
        # testing for viz
        accum = 0
        for r in res :
            accum = accum + r
        df_new.loc['OOD Inhalt in % {}'.format(j)]['{}'.format(i)] = accum / 6


df_new.to_excel('rxrx1_whylogs_results_incremental.xlsx')
```
